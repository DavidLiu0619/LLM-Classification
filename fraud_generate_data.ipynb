```python
# fraud_cls_end2end.py
# ------------------------------------------
# 1) Generate synthetic dataset
import random, os, json, numpy as np, pandas as pd
from datetime import datetime
random.seed(42); np.random.seed(42)

OUT_CSV = "fraud_notes_synth.csv"

first_party_templates = [
    "Customer admits they opened the account and later disputed charges after overspending.",
    "Member used their own card and now claims services not received after a billing dispute.",
    "Caller reports they signed for the loan but says interest was too high and refuses payment.",
    "Customer acknowledges they shared credentials with a friend who made purchases.",
    "Borrower confirms they applied for the card but denies late fees and wants reversal.",
    "Account owner authorized recurring subscription then attempted chargebacks afterward.",
    "Policyholder states they filed the claim but now disputes deductible responsibility.",
    "Client acknowledges mailing address and email are correct; disputes being responsible for charges.",
    "Customer admits enrollment was theirs; now claims transactions were misunderstood.",
    "Cardholder confirms prior relationship with merchant and agreed to terms at checkout."
]

third_party_templates = [
    "Victim reports an unknown account opened using their SSN and a different mailing address.",
    "Card was shipped to an address the customer does not recognize; transactions appear out of state.",
    "Customer denies ever applying; credit pull occurred without consent; police report submitted.",
    "Caller states phone number and email on file are unfamiliar; SIM swap suspected.",
    "Transactions from a device the customer does not own; IP tied to another region.",
    "Employer flagged payroll diversion to a new bank account not set up by employee.",
    "Multiple accounts opened the same day with mismatched KYC details; customer unaware.",
    "Victim reports stolen wallet; merchant attempted delivery to alternate address.",
    "Account takeover alert: password reset via unknown device; contact info changed.",
    "Customer claims identity theft; FTC affidavit filed; card reissued to verified address."
]

# Light variation generators
def vary(note):
    addenda = [
        "", " Investigating with risk team.", " Escalated for manual review.",
        " Reported to authorities.", " Dispute form submitted.", " Case pending documents."
    ]
    swaps = [
        ("customer", "client"), ("unknown", "unrecognized"), ("reports", "states"),
        ("card", "payment card"), ("address", "mailing address"), ("charges", "transactions")
    ]
    s = note
    for a, b in swaps:
        if random.random() < 0.4:
            s = s.replace(a, b)
    s += addenda[random.randint(0, len(addenda)-1)]
    return s

rows = []
n_each = 300   # total ~600 examples; increase if you want
for _ in range(n_each):
    rows.append({"text": vary(random.choice(first_party_templates)), "label": "first_party"})
    rows.append({"text": vary(random.choice(third_party_templates)), "label": "third_party"})

random.shuffle(rows)
df = pd.DataFrame(rows)
df.to_csv(OUT_CSV, index=False)
print(f"[Data] Wrote {len(df)} rows to {OUT_CSV}")
```
